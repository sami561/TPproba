# Rock Paper Scissors

```python
# module random pour le shuffle
import random
# for_time_delay
import time

print("\033[1;30;41m")
print("Welcome to the ðŸ§± ðŸ“„ âœ‚ï¸ game \n")
print("rules :\n"
                                "\tRock vs paper->paper wins \n"
                                "\tRock vs scissor->Rock wins \n"
                                "\tpaper vs scissor->scissor wins \n")

# Display Function


def replay_function():
    print("Do you want to play again? (y/n)")
    replay = input()
    replay = replay.lower()
    if replay == 'y':
        return start_game()
    elif replay == 'n':
        print("\nThanks for playing")
        time.sleep(3)
    return "bye bye"
# apres time.sleep le programme reprend a nouveau et ne se termine pas
# main


def start_game():
    while True:
        player = input("Choose 'r'=rock\t'p'=paper\t's'=scissors : ")
# si le player ecrit r,p,s en majuscule
        player = player.lower()
# loop if player doesn't enter the right keys
        while player != 'r' and player != 'p' and player != 's':
            player = input("enter valid input: ")


# Verification Player Choice Function
        if player == 'r':
            choice = 'rock'
        elif player == 'p':
            choice = 'paper'
        else:
            choice = 'scissor'
    # print player choice
        print("your choice choice is: " + choice)
        print("\npc turn : ....")
        time.sleep(3)  # Sleep for 3 seconds

# computer choice
        pc = random.choice(['r','p','s'])

# replay function
        if pc == 'r':
            pc_choice = 'rock'
        elif pc == 'p':
            pc_choice = 'paper'
        else:
            pc_choice = 'scissor'
        print("Computer choice is: " + pc_choice)
    # winning condition
        if (player == 'r' and pc == 's') or (player == 's' and pc == 'r'):
            print("rock wins")
            game_result = 'rock'
        elif (player == 'r' and pc == 'p') or (player == 'p' and pc == 'r'):
            print("paper wins")
            game_result = 'paper'
        elif (player == 's' and pc == 'p') or (player == 'p' and pc == 's'):
            print("scissor wins")
            game_result = 'scissors'
        # tie
        if player == pc:
            print("it's a tie")
            replay_function()
        # Computer Or Player Win
        if game_result == choice:
            print("!!!!! User wins!!!!!")
        else:
            print("!!!!! Computer wins !!!!!")
        replay_function()


start_game()
```
